```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.set(style='whitegrid')
```

```python
# 1. Bar Plot
plt.figure(figsize=(8, 5))
sns.countplot(data=titanic, x='survived', palette='Set2')
plt.title('Count of Survived vs Not Survived')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
# 2. Histogram
plt.figure(figsize=(10, 5))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=titanic, x='class', y='survived', palette='Set1')
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# 4. Box Plot
plt.figure(figsize=(10, 5))
sns.boxplot(data=titanic, x='survived', y='age', palette='Set3')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
#5. Violin Plot
plt.figure(figsize=(10, 5))
sns.violinplot(data=titanic, x='class', y='age', hue='survived', split=True, palette='muted')
plt.title('Age Distribution by Class and Survival')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
#6. Count Plot
plt.figure(figsize=(8, 5))
sns.countplot(data=titanic, x='sex', hue='survived', palette='pastel')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
#7. Bar Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=titanic, x='sibsp', y='survived', palette='Set2')
plt.title('Survival Rate by Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#8. Pie Chart
plt.figure(figsize=(8, 8))
class_counts = titanic['class'].value_counts()
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Passengers by Class')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```
